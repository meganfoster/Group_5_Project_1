```python
import requests
from pprint import pprint
import pandas as pd

# Read CSV Files with top grossing and best picture winning films
award_csv_path = "Output/newawardwinners.csv"
topgross_csv_path = "Output/Top_grossing_films.csv"
award_movie_df = pd.read_csv(award_csv_path)
topgross_movie_df = pd.read_csv(topgross_csv_path)
award_movie_df=award_movie_df.rename(columns={"entity":"Movie Title"})

api_key = "c847849d"
url = f"http://www.omdbapi.com/?apikey={api_key}&t="
```

```python
award_movie_titles = award_movie_df["Movie Title"]
IMDB_IDs = []
movie_not_found = []

# Run for loop through award movie titles to OMDB API to retrieve IMDB ID
for movie in award_movie_titles:
    try:
        movie_request = requests.get(url + movie).json()
        IMDBID = movie_request["imdbID"]
        IMDB_IDs.append(IMDBID)
    
    except KeyError:
        movie_not_found.append(movie)

print(IMDB_IDs)
print(movie_not_found)       
```

```python
# Make API request to Bechdel API and retrieve Bechdel Scores for award-winning movies
url = "http://bechdeltest.com/api/v1/getMovieByImdbId?imdbid="
award_IMDB_IDs = award_movie_df["IMDB ID"]
bechdel_scores = []

for movie in award_IMDB_IDs:
    query_url = url + movie

    try:
        movie_request = requests.get(query_url).json()
        bechdel_score = movie_request['rating']
        bechdel_scores.append(bechdel_score)
    
    except:
        bechdel_scores.append("Unknown")
print(bechdel_scores)       
```

```python
# Add Bechdel Scores to award movie dataframe
award_movie_df["Bechdel Score"] = bechdel_scores
award_movie_df.to_csv('Output/award_movies_bechdel.csv', index=False, header=True)
award_movie_df
```
