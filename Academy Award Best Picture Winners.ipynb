```python
# Import dependencies
import requests
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Read in csv file & create data frame
filepath = 'Output/awardwinners.csv'
awards_df = pd.read_csv(filepath)
awards_df.head()
```

```python
# Clean data & create new data frame to show only best picture category whose name changed over the years
category_df = awards_df.set_index('category')
category_df = category_df.loc[(['OUTSTANDING PICTURE','OUTSTANDING PRODUCTION','OUTSTANDING MOTION PICTURE',
                              'BEST MOTION PICTURE','BEST PICTURE'])]
category_df
```

```python
# Clean data & create new data frame to show just the winners of best picture for the year
winners_df = category_df.loc[category_df['winner'] == 'Yes']
winners_df = winners_df[['year','entity']]
winners_df.reset_index(inplace = True)
winners_df
```

```python
# Export dataframe to csv file for further analysis
winners_df.to_csv('Output/newawardwinners.csv', index = False)
```
